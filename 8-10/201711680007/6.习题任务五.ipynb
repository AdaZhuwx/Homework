```python
#1、随机生成1万个整数，范围在0-10万之间，分别进行简单选择排序、快速排序（自行递归实现的）以及内置sort函数3种排序，打印出3种排序的运行时间。
import random
import time
def simple_sort(seq):
    start_time=time.time()
    for i in range(0,len(seq)-1):
        min=i
        for j in range(i+1,len(seq)):
            if seq[j]<seq[i]:
                min=j
                seq[i],seq[min]=seq[min],seq[i]
    end_time=time.time()
    print('简单选择排序运行时间为：',end_time-start_time)
    
    
    
def recursive_sort(seq):
    left_seq=[]
    right_seq=[]
    k=seq[0]
    start_time=time.time()
    for number in seq[1:]:
        if number<=k:
            left_seq.append(number)
            
        else:
            right_seq.append(number)
            
    if left_seq:
        left_seq=recursive_sort(left_seq)
        
    if right_seq:
        right_seq=recursive_sort(right_seq)
    
    end_time=time.time()
    print('快速排序运行时间为：',end_time-start_time)
    #运行时会出错，显示超出最大递归深度
    
    
    
def inside_sort(seq):
    start_time=time.time()
    seq.sort()
    end_time=time.time()
    print('内置sort函数运行时间为：',end_time-start_time)
    
    
    
seq=[random.randint(0,100001)for i in range(10000)]
simple_sort(seq)
recursive_sort(seq)
inside_sort(seq)
```

```python
#随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。随机生成1万个整数，范围在0-10万之间，求其中每个整数出现的次数。并按照整数大小排序输出整数及出现次数。

import random

def random_numbers(n):
    numbers=[]
    num=0
    for i in range(n):
        a=random.randint(0,100000)
        numbers.append(a)
    return numbers

def count_nums_freq(numbers):
    nums_freq_pairs=[]
    for number in numbers:
        for item in nums_freq_pairs:#一组item，前一个为词，后一个为词频
            if number==item[0]:
                item[1]+=1
                break
        else:
            nums_freq_pairs.append([number,1])
    nums_freq_pairs.sort()
    return nums_freq_pairs

def main():
    numbers=random_numbers(10000)
    pairs=count_nums_freq(numbers)
    print(pairs)
    
    
main()
```
